```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('C:/Users/shekh/Downloads/social_media_mental_health_dataset.csv')
#  Distribution of users by gender
gender_counts = dataset['Gender'].value_counts()
plt.figure(figsize=(6,4))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'blue', 'green'])
plt.title("Distribution of Users by Gender")
plt.ylabel("")
plt.show()
```

```python
# . Most popular social media platforms
# . Most popular social media platforms
platform_counts = dataset['Social_Media_Platform'].value_counts()
plt.bar(platform_counts.index, platform_counts, color='purple', edgecolor='black')
plt.title("Most Popular Social Media Platforms")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#.How much time do users spend on social media daily?

platform_usage = dataset.groupby('Social_Media_Platform')['Daily_Usage_Hours'].mean()

plt.bar(platform_usage.index, platform_usage, color='teal', edgecolor='black')
plt.title("Average Daily Social Media Usage per Platform")
plt.xlabel("Platform")
plt.ylabel("Average Hours")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
#.How much time do users spend on social media daily?

# Pie chart for average daily social media usage per platform
plt.figure(figsize=(8,6))
plt.pie(platform_usage, labels=platform_usage.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140, shadow=True)
plt.title("Average Daily Social Media Usage per Platform")
plt.show()

```

```python


# . Social media usage by age group
dataset['Age_Group'] = pd.cut(dataset['Age'], bins=[10, 20, 30, 40, 50, 60, 70], 
                              labels=["10-20", "21-30", "31-40", "41-50", "51-60", "61-70"])
age_group_usage = dataset.groupby('Age_Group', observed=False)['Daily_Usage_Hours'].mean()
plt.bar(age_group_usage.index, age_group_usage, color='purple', edgecolor='black')
plt.title("Average Daily Social Media Usage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Hours")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

```python


# . Correlation between social media usage and mental health
# Correlation between social media usage and mental health
correlation = dataset[['Daily_Usage_Hours', 'Mood_Score', 'Anxiety_Score', 'Depression_Score', 'Stress_Score']].corr()

plt.figure(figsize=(8,6))
plt.barh(correlation.columns, correlation['Daily_Usage_Hours'], color='teal')
plt.xlabel("Correlation Coefficient")
plt.ylabel("Mental Health Factors")
plt.title("Correlation Between Social Media Usage and Mental Health")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Count occurrences of each platform
dataset = pd.read_csv('C:/Users/shekh/Downloads/social_media_mental_health_dataset.csv')
platform_counts = dataset["Social_Media_Platform"].value_counts()

# Create a bar chart
plt.bar(platform_counts.index, platform_counts.values, color="purple", edgecolor="black")

# Add labels and title
plt.xlabel("Social Media Platform",fontsize=15)
plt.ylabel("Number of Users",fontsize=15)
plt.title("Most Popular Social Media Platforms",fontsize=15)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Group by Daily Usage Hours and calculate mean mental health scores

usage_vs_mental_health = dataset.groupby("Daily_Usage_Hours")[["Mood_Score", "Anxiety_Score", "Depression_Score", "Stress_Score"]].mean()

# Line plot for mental health impact
plt.figure(figsize=(10, 6))
for col in ["Mood_Score", "Anxiety_Score", "Depression_Score", "Stress_Score"]:
    plt.plot(usage_vs_mental_health.index, usage_vs_mental_health[col], marker="o", linestyle="-", label=col)

plt.xlabel("Daily Social Media Usage (Hours)")
plt.ylabel("Average Score")
plt.title("Impact of Social Media Usage on Mental Health")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Scatter plot: Sleep hours vs. Mental Health Scores

plt.figure(figsize=(10, 6))
plt.scatter(dataset["Sleep_Hours"], dataset["Anxiety_Score"], color="red", alpha=0.5, label="Anxiety Score")
plt.scatter(dataset["Sleep_Hours"], dataset["Depression_Score"], color="purple", alpha=0.5, label="Depression Score")
plt.scatter(dataset["Sleep_Hours"], dataset["Stress_Score"], color="orange", alpha=0.5, label="Stress Score")

# Add labels and title
plt.xlabel("Sleep Hours")
plt.ylabel("Mental Health Scores")
plt.title("Impact of Sleep on Mental Health")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
# Line plot: Exercise Hours vs. Stress Score
exercise_vs_stress = dataset.groupby("Exercise_Hours")["Stress_Score"].mean()


plt.plot(exercise_vs_stress.index, exercise_vs_stress.values, marker="o", linestyle="-", color="blue", label="Stress Score")

# Add labels and title
plt.xlabel("Exercise Hours")
plt.ylabel("Average Stress Score")
plt.title("Impact of Exercise on Stress Levels")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
# 10. Scatter plot: Social media usage vs. Anxiety Score
plt.figure(figsize=(8,5))
plt.scatter(dataset['Daily_Usage_Hours'], dataset['Anxiety_Score'], color='purple', alpha=0.5, edgecolors='black')
plt.title("Daily Social Media Usage vs. Anxiety Score")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Anxiety Score")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 9. Average stress score per country
country_stress = dataset.groupby('Country')['Stress_Score'].mean().sort_values()

plt.bar(country_stress.index, country_stress, color='red', edgecolor='black')
plt.title("Average Stress Score per Country")
plt.xlabel("Country")
plt.ylabel("Average Stress Score")
plt.xticks(rotation=50)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# 11. Most used feature analysis
feature_usage = dataset['Most_Used_Feature'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(feature_usage.index, feature_usage, color='blue', edgecolor='black')
plt.title("Most Used Social Media Features")
plt.xlabel("Feature")
plt.ylabel("Usage Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python


# Pie chart for most used features
plt.figure(figsize=(8,6))
plt.pie(feature_usage, labels=feature_usage.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140, shadow=True)
plt.title("Most Used Social Media Features")
plt.show()
```

```python
# 12. Most stressed age group
age_group_stress = dataset.groupby('Age_Group', observed=False)['Stress_Score'].mean()
plt.bar(age_group_stress.index, age_group_stress, color='orange', edgecolor='black')
plt.title("Average Stress Score by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Stress Score")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Print the most stressed age group
most_stressed_group = age_group_stress.idxmax()
print(f"The most stressed age group is: {most_stressed_group}")
```

```python
# 13. Age group taking therapy sessions
age_group_therapy = dataset.groupby('Age_Group', observed=False)['Therapy_Sessions'].sum()
plt.figure(figsize=(8,5))
plt.bar(age_group_therapy.index, age_group_therapy, color='green', edgecolor='black')
plt.title("Number of People Taking Therapy Sessions by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of People")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the age group taking the most therapy sessions
most_therapy_group = age_group_therapy.idxmax()
print(f"The age group taking the most therapy sessions is: {most_therapy_group}")
```

```python
# 14. Social media platform usage by age group
# 14. Social media platform usage by age group
age_platform_usage = dataset.groupby(['Age_Group', 'Social_Media_Platform'], observed=False).size().unstack()


age_platform_usage.plot(kind='bar', stacked=True, colormap='plasma', edgecolor='black', figsize=(10,6))
plt.title("Social Media Platform Usage by Age Group",fontsize=15)
plt.xlabel("Age Group",fontsize=15)
plt.ylabel("Number of Users",fontsize=15)
plt.xticks(rotation=45)
plt.legend(title="Platform")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# 15. Social media platform usage by gender
# 15. Social media platform usage by gender
gender_platform_usage = dataset.groupby(['Gender', 'Social_Media_Platform'], observed=False).size().reset_index(name='User_Count')

pivot_data = gender_platform_usage.pivot(index='Gender', columns='Social_Media_Platform', values='User_Count')

pivot_data.plot(kind='bar', figsize=(10,6), colormap='plasma', edgecolor='black')
plt.title("Social Media Platform Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.xticks(rotation=0)
plt.legend(title="Platform")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 16. Stress levels by gender
# 16. Stress levels by gender
gender_stress = dataset.groupby('Gender', observed=False)['Stress_Score'].mean()

plt.figure(figsize=(8,5))
plt.pie(gender_stress, labels=gender_stress.index, autopct='%1.2f%%', colors=['purple', 'grey', 'pink'], startangle=90,explode=[0.0, 0.0, 0.1],shadow=True)  # Added a third color

plt.title("Average Stress Levels by Gender")
plt.show()

# Print which gender is more stressed
most_stressed_gender = gender_stress.idxmax()
print(f"The gender with higher average stress levels is: {most_stressed_gender}")

```

```python
# 17. Country with highest average daily social media usage
country_usage = dataset.groupby('Country', observed=False)['Daily_Usage_Hours'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,5))
plt.bar(country_usage.index, country_usage, color='darkblue', edgecolor='black')
plt.title("Average Daily Social Media Usage by Country")
plt.xlabel("Country")
plt.ylabel("Average Daily Usage Hours")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Print the country with the highest usage
most_active_country = country_usage.idxmax()
print(f"The country with the highest average daily social media usage is: {most_active_country}")
```

```python
# 18. Average exercise hours by gender
gender_exercise = dataset.groupby('Gender', observed=False)['Exercise_Hours'].mean()

# Print the result
most_active_gender = gender_exercise.idxmax()
print(f"The gender that exercises more on average is: {most_active_gender}")
```

```python
# 19. Mental health issues by age group
age_mental_health = dataset.groupby('Age_Group', observed=False)[['Stress_Score', 'Anxiety_Score', 'Depression_Score']].mean()
c=['purple','green','orange']
# Plotting
plt.figure(figsize=(10,6))
age_mental_health.plot(kind='bar', figsize=(10,6), color=c, edgecolor='black')
plt.title("Average Stress, Anxiety, and Depression Levels by Age Group",fontsize=15)
plt.xlabel("Age Group",fontsize=15)
plt.ylabel("Average Score",fontsize=15)
plt.xticks(rotation=0)
plt.legend(title="Mental Health Issues")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the most affected age group
most_stressed = age_mental_health['Stress_Score'].idxmax()
most_anxious = age_mental_health['Anxiety_Score'].idxmax()
most_depressed = age_mental_health['Depression_Score'].idxmax()

print(f"Most stressed age group: {most_stressed}")
print(f"Most anxious age group: {most_anxious}")
print(f"Most depressed age group: {most_depressed}")
```
