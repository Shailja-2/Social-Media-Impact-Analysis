```python
#1.How is the age distribution of users?

plt.hist(dataset["Age"], bins=10, color="purple", edgecolor="black", alpha=0.7)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.title("Age Distribution of Users")

# Show the plot
plt.show()
```

```python

#2.What is the gender distribution?
gender_distribution = dataset["Gender"].value_counts()

# Create a pie chart
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%',  startangle=60, colors=['pink', 'skyblue','purple'])

# Add title
plt.title("Gender Distribution of Users")

# Show the plot
plt.show()
```

```python
#3.Which country has the most users in this dataset?

import matplotlib.pyplot as plt


country_counts = dataset["Country"].value_counts()

c =['orange','pink','magenta','blue','purple','black']

plt.bar(country_counts.index, country_counts.values, color=c, edgecolor="black")

plt.xlabel("Country")
plt.ylabel("Number of Users")
plt.title("Number of Users per Country")

plt.xticks(rotation=45)# Rotate x-axis labels for readability

# Show the plot
plt.show()
```

```python
#4.How much time do users spend on social media daily?

time_spend=dataset.groupby("Daily_Usage_Hours")["Stress_Score"].mean()
plt.bar(time_spend.index, time_spend.values, color='skyblue',edgecolor=0.5)
plt.xlabel("Country")
plt.ylabel("Number of Users")
plt.title("Number of Users per Country")
plt.show()
```
